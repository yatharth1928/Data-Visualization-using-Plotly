```python
import pandas as pd
import seaborn as sns
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plotly
import matplotlib.font_manager as font_manager
from datetime import datetime,timedelta
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff
```

```python
#line plot
df=pd.DataFrame({"Years at work" : [1,2,3,4],"income":[6500,10000,12000,14000]})
```

```python
df.head()
```

```python
fig=px.line(df,x="Years at work",y="income",title="Income vs Years at work")
fig.show()
```

```python
stud=['Steve','Leo','Chaf','Capri']

maths=[89,78,90,98]
science=[78,96,67,87]
programming = [81,87,92,99]
data_base_subject=[91,93,67,89]
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=stud,y=maths,name='Maths marks',line=dict(color='blue',width=4)))
fig.add_trace(go.Scatter(x=stud,y=science,name='Science',line=dict(color='red',width=4)))
fig.add_trace(go.Scatter(x=stud,y=programming,name='Programming',line=dict(color='green',width=4)))
fig.add_trace(go.Scatter(x=stud,y=data_base_subject,name='Data Base',line=dict(color='yellow',width=4)))
fig.update_layout(title='Student Marks',xaxis_title='Student Name',yaxis_title='Marks')
```

**Bar Plot**

```python
df=sns.load_dataset('titanic')
```

```python
df.head(10)
```

```python
fig=px.bar(df,x='pclass',y='fare',color='sex',title='Fare vs Pclass')
fig.show()
```

**Area Chart**

```python
fig=px.area(df,x='sex',y='age',color='survived',title='Area plot for survival based on sex vs age')
fig.show()
```

**Stacked area chart**

```python
stud=['Rave','Leo','Steve']
fig=go.Figure()
fig.add_trace(go.Scatter(x=stud,y=[40,10,20], hoverinfo='x+y',mode='lines',line=dict(width=0.6,color='rgb(172,89,200)'),stackgroup='one'))
fig.add_trace(go.Scatter(x=stud,y=[50,5,2], hoverinfo='x+y',mode='lines',line=dict(width=0.6,color='rgb(100,51,190)'),stackgroup='one'))
```

**Pie Chart**

```python
stud=['Leo','Steve','Claire','Caprio','Sharma','Kiya']
marks=[34,56,76,87,45,90,91]

fig=go.Figure(data=[go.Pie(labels=stud,values=marks,name='Marks',pull=[0,0,0,0,0.2,0,0])])
fig.update_traces(textposition='inside',textinfo='percent+label')
fig.show()
```

**Multilevel pie chart**

```python
stud=['Leo','Steve','Claire','Caprio','Sharma','Kiya']
teacher=['Steve',"","Caprio","Leo","Steve","Sharma"]
marks=[34,56,76,87,45,90,91]

fig=go.Figure(go.Sunburst(labels=stud,parents=teacher,values=marks,))
fig.update_layout(margin=dict(t=1,l=1,r=1,b=1))
fig.show()
```
